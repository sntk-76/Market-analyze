## section 8 

In this analysis, we are working with financial data represented in the form of a DataFrame. We aim to identify and remove noisy data points that could distort our analysis and visualizations.

### Identifying Noise

We observe that there is noise in the charts, particularly in the "Open vs Close" and "High vs Low" charts. This noise is characterized by unusual data points that do not conform to the expected patterns of financial data.

For instance, in the "High vs Low" chart, we notice a data point where the price is abnormal. Specifically, we are looking for data points where the following conditions are met:

- The 'High' price is greater than 4000.
- The 'High' price is less than 8000.
- The 'Low' price is less than 2000.

### Noise Removal

To address this issue, we implement a data cleaning procedure in Python. Here's how it works:

1. We identify the specific rows (data points) that meet our noise conditions using boolean indexing. These rows represent the data points we want to remove.

2. We use the `drop` method to remove these rows from our DataFrame, ensuring that the DataFrame is updated accordingly.

### Visualization

After removing the noisy data points, we recreate a scatter plot to visualize the effect of the noise removal. This will help us determine if our implementation was successful in cleaning the data and improving the quality of our visualizations.


```python
index_to_remove = df[(df['High'] > 4000) & (df['High']< 8000) & (df['Low'] < 2000)].index
df.drop(axis=0,index=index_to_remove,inplace=True)
plt.scatter(df['High'],df['Low'])
plt.xlabel('High price')
plt.ylabel('Low price')
plt.title('High price vs Low price')
```

```python
index_to_remove = df[(df['Open'] > 4000) & (df['Open']< 6000) & (df['Close'] < 2000)].index
df.drop(axis=0,index=index_to_remove,inplace=True)
plt.scatter(df['Open'],df['Close'])
plt.xlabel('Open price')
plt.ylabel('Close price')
plt.title('Open price vs Close price')
```

## section 9

In this section, we focus on identifying and removing noisy data from our dataset. Our analysis reveals that there are certain records with zero trading volume for several years. We consider these data points as noise and decide to eliminate them from our dataset.

## Identifying Noise

We first identified noisy data points by examining charts that display relationships between "Date" and "Volume." In these charts, we noticed that there are instances where the trading volume is consistently zero for extended periods. This observation led us to conclude that these data points are not representative of typical trading behavior and can be considered outliers.

We observed this trend not only in the "Date" vs "Volume" chart but also in other charts such as "Close" vs "Volume" and "Open" vs "Volume." In each case, we found that the presence of zero trading volume for extended periods indicated the need to clean the data.

## Data Cleaning

To remove the noisy data from our dataset, we implemented the following steps:

1. **Identify the Starting Point:** We iterated through the "Volume" column to find the index of the first non-zero trading volume, which we stored in the variable `first_volume`.

2. **Filter the Data:** We filtered the DataFrame to keep only the data points from `first_volume` onwards. This effectively removed the records with zero trading volume for extended periods.



```python
for index,row in enumerate(df['Volume']):
    if row > 0 :
        first_volume = index
        break

print(df.iloc[first_volume-1])
print('\n')
print(df.iloc[first_volume])    
df = df.iloc[first_volume:]
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['Date'],df['Volume'])
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Stock Volume Over Time')
plt.xticks(rotation = 90)
```

```python
df = df[df['Volume']<0.9*1e10]
plt.scatter(df['Open'],df['Volume'])
plt.xlabel('Open price')
plt.ylabel('Volume')
plt.title('Openning price vs Volume')
plt.xticks(rotation = 90)
```

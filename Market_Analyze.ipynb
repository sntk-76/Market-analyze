1. As the first step, we should import the libraries that needs to be imported in project: 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2. The second step is reading the csv file that contains the data that are needed for our project : 

```python
csv_file = pd.read_csv('Market.csv')
df = pd.DataFrame(csv_file)
df
```

3. In this project we just need to convert all the numerical columns to INT type, so we loop over the dataframe and convert all FLOAT types to INT types : 

```python
for col in df.columns:
    for index, cell in enumerate(df[col]):
        try:

            df.at[index, col] = int(cell)
        except (ValueError, TypeError):

            continue
df        
```

4. There are some markets that are available in our data frame but here we just want to analyze the NYA market, so we use one of the built-in capabilities in pandas data frame and extract the rows that record the NYA market :

```python
df = df[df['Index']=='NYA']
df
```

5. Then we want to create a linear plot that shows us the "Adj Close" according to the time, actually this plot shows us a linear relation of the price over time :

```python
df['Date']= pd.to_datetime(df['Date'])
plt.figure(figsize=(12,8))
plt.plot(df['Date'],df['Adj Close'])
plt.xlabel('Date')
plt.ylabel('Adj close price')
plt.title('Daily time frame')
plt.xticks(rotation=90)
plt.show()
```
